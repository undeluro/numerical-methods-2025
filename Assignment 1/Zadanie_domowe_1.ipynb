## Wskazówki
### Wybrane własności macierzy
1. $A^{-1}A = I$, dla każdej nieosobliwej macierzy kwadratowej.
1. $Q^T = Q^{-1}$, jeśli Q jest macierzą ortonormalną.
1. $(AB)^T=B^TA^T$, dla dowolnych macierzy.
1. $(AB)^{-1}=B^{-1}A^{-1}$, dla dowolnych macierzy.
